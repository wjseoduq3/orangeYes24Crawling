```python
all_title = []
all_auth = []
all_pubDate = []
all_price = []
pageNum = int(input("페이지 수를 입력하세요 : "))

for page in range(1,pageNum+1):
    url="https://www.yes24.com/24/Category/Display/001001009003?PageNumber=" + str(page)
    my_header = {'User-Agent':'Mozilla/5.0' }
    html = requests.get(url, headers=my_header)
    page = bs(html.text, 'lxml')
    tag_div = page.select('div.goods_info')
    title_div = page.select('div.goods_name')

    for i in range(len(tag_div)):
        print(f"{page}페이지의 {i}번째 도서 검색 중")

        second_element_text = title_div[i-1].get_text()  # 세번째 - 1 적용해야 함
        title = second_element_text.split('\n')[2]

        # title = tag_div[i].select('div > a',{'class':'goods_name'})[0].text
        # print(title)
        # title = tag_div[i].find_all('div',{'class':'goods_name'}).find_all('a')[0].text

        auth = tag_div[i].find_all('span',{'class':'goods_auth'})[0].text.replace('\n', '').replace('\r','').replace('저','').strip()
        pubDate = tag_div[i].find('span',{'class':'goods_date'}).text
        price = tag_div[i].find('em',{'class':'yes_b'}).text + '원'

        all_title.append(title)
        all_auth.append(auth)
        all_pubDate.append(pubDate)
        all_price.append(price)

        time.sleep(1)

df = pd.DataFrame({'title':all_title, 'auth':all_auth, 'pubDate':all_pubDate, 'price':all_price})
df.to_csv('yes24.csv', encoding='utf-8-sig', index=False)
df
```
